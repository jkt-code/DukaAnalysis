```python
# Lets extract transaction cost table from safaricom webpage
import requests
from bs4 import BeautifulSoup

url = "https://www.safaricom.co.ke/main-mpesa/m-pesa-for-you/tariffs-limits/consumer-tariffs-limits"

response = requests.get(url)
soup = BeautifulSoup(response.content,'html.parser')

table = soup.find('table')

transCost_SendMoney = pd.read_html(str(table))[0]
transCost_SendMoney 
```

```python
transCost_SendMoney.columns = transCost_SendMoney.iloc[1]
transCost_SendMoney = transCost_SendMoney[2:].reset_index(drop=True)
transCost_SendMoney = transCost_SendMoney.replace('Free', 0)
transCost_SendMoney = transCost_SendMoney.fillna(0)
transCost_SendMoney 
```

```python
# Lets extract transaction cost for paying via paybill
import requests
from bs4 import BeautifulSoup

url1 = "https://www.tuko.co.ke/facts-lifehacks/guides/441623-m-pesa-paybill-charges-how-cost-pay-bills/"

response1 = requests.get(url1)
soup1 = BeautifulSoup(response1.content,'html.parser')

table1 = soup1.find_all('table')

transCost_Paybill = pd.read_html(str(table1))[1]
transCost_Paybill
```

```python
# Use the first row as column headers 
transCost_Paybill.columns = transCost_Paybill.iloc[0] 
transCost_Paybill = transCost_Paybill[1:].reset_index(drop=True)
transCost_Paybill
```

```python
# Split the 'Bands (KES)' column into 'min' and 'max' columns
transCost_Paybill[['min', 'max']] = transCost_Paybill['Bands (KES)'].str.split('', expand=True)

# Trim any leading/trailing whitespace and convert to numeric
transCost_Paybill['min'] = transCost_Paybill['min'].str.replace(',', '').str.strip().astype(int)
transCost_Paybill['max'] = transCost_Paybill['max'].str.replace(',', '').str.strip().astype(int)

# Reorder columns to place 'min' and 'max' to the right of 'Bands (KES)'
cols = list(transCost_Paybill.columns)
bands_index = cols.index('Bands (KES)')
new_cols = cols[:bands_index + 1] + ['min', 'max'] + cols[bands_index + 1:-2]
transCost_Paybill = transCost_Paybill[new_cols]
transCost_Paybill
```

```python
# Ensure MIN and MAX columns are numeric in transCost_SendMoney
transCost_SendMoney['MIN'] = transCost_SendMoney['MIN'].astype(float)
transCost_SendMoney['MAX'] = transCost_SendMoney['MAX'].astype(float)

transCost_Paybill['min'] = transCost_Paybill['min'].astype(float)
transCost_Paybill['max'] = transCost_Paybill['max'].astype(float)
```

```python
def find_costNumber(value):
    for index, row in transCost_SendMoney.iterrows():
        if row['MIN'] <= value <= row['MAX']:
            return row['TRANSFER TO M-PESA USERS, POCHI LA BIASHARA AND BUSINESS TILL TO CUSTOMER']
    return None

def find_costPaybill(value):
    for index, row in transCost_Paybill.iterrows():
        if row['min'] <= value <= row['max']:
            return row['Customer fee (KES)']
    return None 

# Create a new column in transmodePurchases to store the transaction cost
transmodePurchases['transcost'] = None

for index, row in transmodePurchases.iterrows():
    if row['transmode_mpesa'] == 'number':
        transmodePurchases.at[index, 'transcost'] = find_costNumber(row['total_spent'])
    elif row['transmode_mpesa'] == 'till':
        transmodePurchases.at[index, 'transcost'] = 0
    elif row['transmode_mpesa'] == 'paybill':
        transmodePurchases.at[index, 'transcost'] = find_costPaybill(row['total_spent'])
    else:
        transmodePurchases.at[index, 'transcost'] = 0
```
